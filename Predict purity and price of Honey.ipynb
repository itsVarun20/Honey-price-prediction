```python
numerical_variables = df.dtypes[df.dtypes == 'float64'].index
Categorical_variables = df.dtypes[df.dtypes == 'object'].index
```

```python
numerical_variables
```

```python
plt.figure(figsize = (15, 20))

index = 1
for  variable in (numerical_variables):
    plt.subplot(5, 2, index)
    sns.histplot(x = df[variable], kde=True)
    plt.title(variable)
    index += 1
    plt.xlabel('')

plt.suptitle('Distribution of Numerical Variables', size=20, weight='bold', y=0.95)
plt.show()
```
